```python
import pandas as pd
df=pd.read_csv("store_customers.csv")
columns=df.columns
print(columns)
```

```python
customer_data=df.dropna()
customer_data=customer_data.rename(columns={"Spending Score (1-100)":"spendingScore"})
customer_data=customer_data.rename(columns={"Annual Income (k$)":"annualIncome"})
```

```python
from sklearn.preprocessing import LabelEncoder  
le=LabelEncoder()
customer_data["Gender"]=le.fit_transform(customer_data["Gender"])
customer_data["Gender"]
```

```python
from sklearn.model_selection import train_test_split
annual_income=customer_data["annualIncome"]
feature_data=customer_data.drop("annualIncome", axis=1)
X_train, X_test, y_train, y_test = train_test_split(feature_data,annual_income, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
import tensorflow as tf

model = tf.keras.models.Sequential([
    
    tf.keras.layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    
    tf.keras.layers.Dense(32, activation='relu'),
    
    tf.keras.layers.Dense(1)   # output layer for income prediction
])
```

```python
model.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mae']
)
```

```python
history = model.fit(
    X_train,
    y_train,
    epochs=50,
    batch_size=32,
    validation_split=0.2
)
```

```python
loss, mae = model.evaluate(X_test, y_test)
```

```python
print("Mean Absolute Error:", mae)
```

```python
predictions = model.predict(X_test)
print(predictions[:5])
```
