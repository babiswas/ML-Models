```python
columns=["area","parking","bedrooms","stories","price"]
subset=df[columns]
corr = subset.corr()
import matplotlib.pyplot as plt

corr['price'].drop('price').plot(kind='bar')

plt.title("Correlation with Price")
plt.ylabel("Correlation")
plt.show()
```
